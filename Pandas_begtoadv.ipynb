```python
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

```python
df=pd.DataFrame(data,index=labels)
df
```

```python
## Calculate the total visits
df["visits"].sum()
```

```python
###  Calculate the average visits
df["visits"].mean()
```

```python
## calculate the mean age for each type of animal
df.groupby('animal')["age"].mean()
```

```python
### Calculate the number of each type of animal
df.groupby("animal")["animal"].count()
```

```python
### Add a new row k to df to snake,4,3,no

df.loc['k']=["snake",4,3,"no"]
```

```python
### Sort according to age(descending)
df.sort_values(["age"],ascending=False)
```

```python
### Sort age in descending order and visit in ascending
df.sort_values(["age","visits"],ascending=[False,True])
```

```python
### Change the priority column yes to 1 and no to 0

df["priority"]=df["priority"].replace({"yes":1,"no":0})
```

```python
df
```

```python
###  Change snake to Python
df["animal"]=df["animal"].replace('snake',"python")
```

```python
### For each animal type and each number of vists find the mean age
df.pivot_table(index="animal",columns="visits",values="age",aggfunc="mean")
```
